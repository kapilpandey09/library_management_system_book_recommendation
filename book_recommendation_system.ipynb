# **Populartiy Based Recommender System**

```python
ratings.merge(books,on="ISBN").shape
```

```python
rating_books = ratings.merge(books,on="ISBN")
```

```python
rating_books.shape
```

```python
rating_books.columns
```

```python
rating_books['Book-Title'].nunique()
```

```python
num_rating_df = rating_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df.head()
```

```python
rating_books['Book-Rating'] = pd.to_numeric(rating_books['Book-Rating'], errors='coerce')
```

```python
# Convert 'Book-Rating' to numeric, replacing non-numeric values with NaN
rating_books['Book-Rating'] = pd.to_numeric(rating_books['Book-Rating'], errors='coerce')

# Now calculate the average rating, excluding NaN values
avg_rating_df = rating_books.groupby('Book-Title')['Book-Rating'].mean(numeric_only=True).reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df.head()
```

```python
avg_rating_df
```

```python
num_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

# **Collaborative Filtering Based Recommender System**

```python
X = rating_books.groupby('User-ID').count()['Book-Rating'] >200
```

```python
high_rating = X[X].index
```

```python
filter_rating = rating_books[rating_books['User-ID'].isin(high_rating)]
```

```python
y = filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(pt).shape
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
  index = np.where(pt.index==book_name)[0][0]
  distances = similarity_score[index]
  book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in book_list:
    print(pt.index[i[0]])

```

```python
recommend('1984')
```
